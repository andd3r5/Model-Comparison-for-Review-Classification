```
import pandas
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (10,6)  # Increase default size of plots

#Accuracy
MultNB_good = {'Training_Volume': [568, 2842, 14210, 71053, 355266, 511584],
          'Accuracy': [64, 64, 66, 70, 77, 79]}

MultNB_bad = {'Training_Volume': [568, 2842, 14210, 71053, 355266, 511584],
          'Accuracy': [64, 64, 64, 64, 67, 68]}

BernoulliNB = {'Training_Volume': [568, 2842, 14210, 71053, 355266, 511584],
          'Accuracy': [64, 64, 64, 65, 68, 69]}

LinearSVC = {'Training_Volume': [568, 2842, 14210, 71053, 355266, 511584],
          'Accuracy': [64, 68, 73, 78, 83, 84]}

RoBERTa = {'Training_Volume': [568, 2842, 14210],#, 71053, 355266, 511584],
          'Accuracy': [64, 72, 75]}#, 0, 0, 0]}

plt.plot(MultNB_good['Training_Volume'], MultNB_good['Accuracy'], label = 'MultinomialNB')
plt.plot(MultNB_bad['Training_Volume'], MultNB_bad['Accuracy'], label = 'MultinomialNB Tf-idf')
plt.plot(BernoulliNB['Training_Volume'], BernoulliNB['Accuracy'], label = 'BernoulliNB')
plt.plot(LinearSVC['Training_Volume'], LinearSVC['Accuracy'], label = 'LinearSVC')
plt.plot(RoBERTa['Training_Volume'], RoBERTa['Accuracy'], label = 'RoBERTa-base')

xmin, xmax, ymin, ymax = plt.axis()
ymin = 60
plt.axis([xmin, xmax, ymin, ymax])
#[plt.axvline(x) for x in [568, 2842, 14210, 71053, 355266, 511584]]

plt.ylabel('Test accuracy in percentage')
plt.xlabel('Number of training observations')
plt.title('Comparison of Test Accuracy')
plt.legend()
plt.show()
```
